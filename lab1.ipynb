```python
file_name = "poem.txt"
sum_of_each_line= 0
no_of_line=0

with open(file_name, 'r') as f:
    for line in f:
        no_of_line +=1
        sum_of_each_line += len(line)
print("sum->",sum_of_each_line)
print("Mean-> ",sum_of_each_line/no_of_line)
```

```python
# group_by_cyl=df.groupby(by=["cyl"]).sum()
# print(group_by_cyl['cty'])
MpgByCyl = []

for i in cylinders: # iterate over all the cylinder levels
    sumpg = 0
    typecount = 0
    for d in mpg: # iterate over all dictionaries
        ##ADD REMAINING CODE
        if d['cyl']==i:
            sumpg+=int(d['cty'])
            typecount+=1
    MpgByCyl.append(sumpg/typecount)
    print(i,sumpg/typecount)
```
